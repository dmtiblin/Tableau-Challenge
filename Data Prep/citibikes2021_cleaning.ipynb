```python
with ZipFile("202101-citibike-tripdata.csv.zip") as myzip:
    data = myzip.open("202101-citibike-tripdata.csv")

#Now, we can read in the data
df_202101 = pd.read_csv(data)
df_202101.dropna(axis=0, inplace=True)

df_202101
```

```python
df_202101.drop(["bikeid","tripduration"], axis =1, inplace = True)
```

```python
df_2021 = pd.concat(
    [df_202111.sample(frac = .2, axis = 0), df_202110.sample(frac = .2, axis = 0),df_202109.sample(frac = .2, axis = 0), df_202108.sample(frac = .2, axis = 0),df_202107.sample(frac = .2, axis = 0),df_202106.sample(frac = .2, axis = 0),df_202105.sample(frac = .2, axis = 0),df_202104.sample(frac = .2, axis = 0),df_202103.sample(frac = .2, axis = 0),df_202102.sample(frac = .2, axis = 0)], axis = 0)

df_2021 = df_2021.rename(columns = {'started_at':'starttime','ended_at':'stoptime', 'start_station_name':'start station name',"start_station_id":"start station id",
                          "end_station_name":"end station name", "end_station_id":"end station id","start_lat":"start station latitude", "start_lng":"start station longitude","end_lat":"end station latitude", "end_lng":"end station longitude", "member_casual": "usertype"})
df_2021.drop('ride_id', axis = 1, inplace = True)
```

```python

df_2021 = pd.concat([df_2021, df_202101.sample(frac = .2, axis = 0)], axis =0)
df_2021.reset_index(drop= True,inplace= True)
```
